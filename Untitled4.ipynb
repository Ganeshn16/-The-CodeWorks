```python
from IPython.display import display
from datasets import load_dataset
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from diffusers import StableDiffusionPipeline
import torch
```

```python
# Load the Mobile UI Design dataset
dataset = load_dataset("mrtoy/mobile-ui-design")

# Print a sample from the dataset to verify
print(dataset['train'][0])  # Print the first item from the training set
```

```python
# Load GPT-2 Model and Tokenizer
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")

# Load Stable Diffusion pipeline for generating mobile UI design images
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")  # Use GPU if available
```

```python
def generate_ui_design(input_text):
    # Tokenize and generate text with GPT-2
    inputs = tokenizer(input_text, return_tensors="pt")
    outputs = model.generate(**inputs, max_new_tokens=100)  # Increase max_new_tokens
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Print the generated text for validation
    print("Generated text:", generated_text)

    # Create a descriptive image prompt
    image_prompt = f"Create a mobile app UI design for: {generated_text.strip()}. Please use English."

    # Generate image using Stable Diffusion
    image = pipe(image_prompt).images[0]

    return image, image_prompt
```

```python
user_input = "Generate a mobile app UI design for a login screen in English."
image, image_prompt = generate_ui_design(user_input)
display(image)
```
